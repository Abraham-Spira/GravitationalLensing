# Introduction
On the 25th of November, 1915, the world of physics was completely transformed. This was the day that Albert Einstein published his General Theory of Relativity. This theory provided an entirely new model of theoretical physics, a model that seemingly contradicted all other generally accepted theories of the time. One of the most notable parts of Einsteins General Theory of Relativity was its total diversion from Newton's theory of gravity. Einstein postulated that gravity was not a force like Newton had theorized, but rather a phenomenon intrinsic to the fabric of the universe, namely the bending of spacetime itself.

The implications of this theory were far and wide, however arguably the most important conclusion that could be made was that gravity, being not a force but rather the curvature of spacetime, could theoretically affect massless particles. According to Newton's equations gravity acted as an attraction between two masses, however Einstein's theory of gravity stated that this perceived force is simply the result of one body traveling along spacetime curved by the other. This means that only one body must have mass for the other's trajectory to be affected by it. This led to the logical yet revolutionary conclusion that mass not only bends the path of other massive bodies, but even of light itself. This bending of light, which soon came to be called gravitational lensing, seemed illogical to many of the great physics minds of the time, however its measurement soon became one of the first proofs for Einstein's theory.

In Einstein's paper theorizing this new concept of gravity he included a set of differential equations, known as Einstein's field equations, which provide the mathematical guidelines for how spacetime bends. However, despite Einstein being the first to write these equations, he was not the first to solve them. That title goes to a fellow German physicist named Carl Schwarzschild. His solution to Einstein's field equations, known as the Schwarzschild metric, was limited to Schwarzschild spacetime (a theoretical spacetime in which the mass' electric charge, angular momentum, and universal cosmological constant are all zero), however it opened up for the opportunity for the trajectories of massive bodies and light particle to be calculated using Einstein's theory. However, in order to calculate the trajectory of light particles in a more realistic fashion, an additional equation is needed, namely the Eikonal equation. This equation is undoubtedly the most important equation in ray optics, and "itâ€™s solutions describe the paths of light rays through complicated media, including the phenomenon of gravitational lensing" (Nolte, 2019).

Additionally, solutions to the Schwarzschild metric reveal two important values, the Schwarzschild radius and the photon sphere. The Schwarzschild radius, calculated using the formula R = 2GM/c^2, is the radius of a massive body as which the curvature of space time is so severe that not even light can escape the gravitational pull. This value is incredibly important because any body where the Schwarzschild radius is larger than the radius of the body itself is deemed a black hole. The photon sphere, which lies at 1.5 times the Schwarzschild radius, is the exact radius at which light is able to maintain a circular orbit. In other words, the photon sphere is the innermost stable circular orbit (ISCO) of a photon. This value is always the same multiple of the Schwarzschild radius by nature of light having a constant speed.

# Computational Approach
In order to calculate the trajectory of photons, the Schwarzschild field equations and the Eikonal equation for ray optics must be reconciled with one another. In order to do this one must solve the Schwarzschild geometry equation for the movement of light: 

$$  ds^{2} = g_{00}c^{2}dt^{2} + g_{rr}dr^{2} = 0$$

Additionally, Eikonal's equation, when rewritten in the terms of Schwarzschild's field equations, provides an effective refractive index of warped Schwarzschild spacetime. This equation, written as a function of a photon's distance from a massive body, is as follows:

$$ n(r) = \frac{1}{-g_{00}(r)} = \frac{1}{1-\frac{2GM}{c^{2}r}} $$

By solving Schwarzschild's differential equation for the change in the velocity of light (velocity of light can change because velocity is a vector, the speed of light, however, is constant), and plugging that solution into Eikonal's equation, one can find the four basic differential equations which describe the movement of light particles in a 2-dimensional plane as a function of their distance from a gravitational body (represented by "r") and the effective refractive index of their current location (represented by "n"). These equations are as follows:

$$ \frac{dx}{ds} = \frac{v_{x}}{n} $$

$$ \frac{dy}{ds} = \frac{v_{y}}{n} $$

$$ \frac{dv_{x}}{ds} = -n^{2}\frac{2GMx}{c^{2}r^{3}} $$

$$ \frac{dv_{y}}{ds} = -n^{2}\frac{2GMy}{c^{2}r^{3}} $$

Using python, we are able to model these differential equations (and the corresponding photon paths) for a variety of photons at different distances from a massive body. In this case, for the sake of simplification and more clear data visualization, stand-in numbers were used for the Schwarzschild radius of the black hole. This number (in this case 10) is unitless and can be understood as simply a scalar value representing the relative size of the black hole. Similarly, the distance values are unitless and are simply there to represent how a photon's path changes in proportion to its distance from a massive body.

There were three main steps within the python code that resulted in the formation of one figure with all the photon paths graphed. The first of which was to define the effective refraction equation as a function of the x and y coordinates of the photon (also allowing for the calculation of their distance from the massive body). Secondly, the flow functions (i.e. Schwarzschild's field equations transformed to represent light pathways, as shown above) were defined (using the results of the refraction equation) in order to continually calculate the rate of change of the directions of the photons. Thirdly, a loop was created to continually integrate the flow function for various original distance values. Lastly, all of these paths were plotted on a graph alongside two shapes. The first of which is a filled-in circle representing the massive body itself, with the circle's radius because equivalent to the body's theoretical Schwarzschild radius. The second shape is the outline of a circle which represents the location of the photon ISCO (photon sphere), located at 1.5 times the Schwarzschild radius.

The code and the resulting graph is shown below:

```python
### Abraham (Bram) Spira
### Mr. Payne, Computational Physics
### Final Project Code - Simulated Photon Paths
### Code based on a the published code of David D. Nolte
### May 22, 2023

import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

plt.close('all') # Closing all previous graphs

# Define constants and variables
l = 10 # Schwarzschild radius of black hole, equal to 2GM/c^2, not true value but put as low whole number for purpose of simulation

# Setting graphing parameters
figure, axes = plt.subplots()
plt.rcParams['figure.figsize'] = [10, 10]

def refraction(x, y): # creating a function to find the refraction index values
    
    r = np.sqrt(x**2 + y**2)
    
    ri = 1/(1 - (l/(r))) # calculation of effective refractive index of Schwarzschild spacetime
    cor = np.abs(1 - 9*(l/r)**(2/8)) # approximation of correction of Eikonal equation, provided by David D. Nolte
    
    dx = ((-cor*(ri**2)*l*x)/((r)**3)) # represents the derivative of movement in x direction
    dy = ((-cor*(ri**2)*l*y)/((r)**3)) # represents the derivative of movement in x direction
    
    return [ri, dx, dy]

def flowfunc(x_y_z_w, dspan): # creating the differential equation representing the flow of photons
    x, y, z, w = x_y_z_w
    [ri,dx,dy] = refraction(x,y)
    
    yi = np.zeros(shape=(4,))
    
    yi[0] = z/ri # assigns first zero to the movement in x direction
    yi[1] = w/ri # assigns second zero to the movement in y direction
    yi[2] = dx # assigns third zero to the derivative of movement in x direction
    yi[3] = dy # assigns fourth zero to the derivative of movement in y direction
    
    return yi # returns the function/atrix representing the four characteristics of photon movement in four dimentions

for v in range(-5,20):
    
    x0 = -100 # first x value to be computed
    y0 = 5*v # series of y values to be computed, starts at -25 and increases in steps of 2
    
    [ri, dx, dy] = refraction(x0,y0) # finds the solutions of the refraction equation for the specific loop iteration
    
    conditions0 = [x0, y0, ri, 0] # sets the initial conditions for the differential equation
    
    tSpan = np.linspace(1, 400, 2000) # creates a series of evenly spaced numbers from 1 to 400 with 2000 numbers
    
    solution = integrate.odeint(flowfunc, conditions0, tSpan) # integrating the flow function for all the pertinent values
    
    xPoints = solution[1:2000, 0] # finds the x coordinates for all pertinent points (the 0 means it is taking the first value from the solution function)
    yPoints = solution[1:2000, 1] # finds the y coordinates for all pertinent points
    
    lines = plt.plot(xPoints, yPoints) # plotting the solutions
    plt.setp(lines, linewidth=1)
    axes.set(xlim=(-100, 100), ylim=(-100, 100)) # setting the axes sizes
    axes.set_aspect(1) # setting the aspect ratio of the grpah to 1:1
    plt.title('Photon Orbits') # titling the graph
    plt.show

# Creating graph of black hole and Schwarzschild Radius
circle = plt.Circle((0, 0), 10) # creating black hole itself
perimeter = plt.Circle((0, 0), 15, fill = False) # creating photon sphere circle

axes.set_aspect(1)
axes.add_artist(circle)
axes.add_artist(perimeter)
plt.title('Photon Orbits')
plt.show()
```

# Results and Discussion
Both graphs clearly showcased curved photon paths, providing clear evidence that, in Schwarzschild spacetime, photons (and light rays accordingly) due bend in the presence of massive bodies. However, despite this notion being already widely known, these graphs provided a clear and easy-to-understand visualization of this phenomenon. As shown in the first graph, the farther the distance a photon is from a massive body, the less its path is curved, something synonymous with the ideas of Newtonian gravity in regards to the distance between two massive bodies. Moreover, the behavior of the photons as they near the Schwarzschild radius of the massive body clearly demonstrates the limitations in the mathematics, as in this simulation once they cross the radius they simply follow a straight line directly to the center. This is because having a distance less than the Schwarzschild radius creates a singularity in the differential equations, and the behavior of light (or anything else) after is passes the Schwarzschild radius is still unknown.

Additionally, the light bending in the first graph can be seen as a simulation of the possibility for misleading cosmic observations due to gravitational lensing. For example, an observer standing at the coordinates (-100, -88) could look towards the positive x and positive y direction, however the true location of whatever they seem might be closer to the point (-50, 73). This is because the presence of a supermassive body is causing the light to bend into the observer's eyes, with this exact case being shown by the middle dark blue photon path in the first graph.

The second graph also provides a clear visualization of how light behaves as it nears the ISCO radius. This radius (also called a photon sphere), as said above, is the only radius at which light can maintain a stable circular orbit. The simulated photon paths in the second graph clearly show that any photon that crosses this radius ultimately spirals into the singularity, whereas any photon that never crosses the photon sphere eventually diverges outwards. This shows that a theoretical photon which falls right on the photon sphere can maintain a circular orbit, as it never crosses over the radius but also never diverges outwards. This second graph also demonstrates that the photon sphere does indeed lie at 1.5 times the Schwarzschild radius when photon paths are calculated according to the Schwarzschild metric, as this is the location at which the circle is graphed.
